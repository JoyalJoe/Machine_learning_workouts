```python
df1 = df.drop(columns= ['RowNumber','CustomerId'], axis=1)
```

df1.isnull().sum()

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dummy=['Gender','Geography','Surname']
for i in dummy:
    df1[i]=le.fit_transform(df1[i].astype(str))
df1   
```

```python
df1.value_counts
```

```python
df1.shape
```

```python
df1['CreditScore'] = df1['CreditScore'].fillna(df1['CreditScore'].mean())
df1['EstimatedSalary'] = df1['EstimatedSalary'].fillna(df1['EstimatedSalary'].mean())
df1['NumOfProducts'] = df1['NumOfProducts'].fillna(df1['NumOfProducts'].mean())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
train,test=train_test_split(df1,test_size=0.2)
```

```python
train
```

```python
test
```

```python
df1.isnull().sum()
```

```python
#data preparastion
trainx=train.iloc[:,0:12]
trainy=train.iloc[:,12]
testx=test.iloc[:,0:12]
testy=test.iloc[:,12]
```

```python
from sklearn.neighbors import KNeighborsClassifier as  KNC
```

```python
knn_model = KNC(n_neighbors=5)
predknn=knn_model.fit(trainx, trainy).predict(testx)
```

```python
predknn
```

```python
testy
```

```python
mean=np.mean(predknn==testy)
```

```python
mean
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import precision_score
```

```python
accuracy_test = precision_score(predknn, testy)
accuracy_test
```

```python
a=[]
for i in range(3,50,2):
    knn_model = KNC(n_neighbors=i)
    predknn=knn_model.fit(trainx, trainy).predict(testx)
    knnacc=np.mean(predknn==testy)
    a.append([i,knnacc])
   
```
