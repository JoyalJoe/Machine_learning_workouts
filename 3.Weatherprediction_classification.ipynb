## Handling missing values:

```python
df.head(5)
```

```python
df.nunique()
```

```python
features_with_na=[features for features in df.columns if df[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(), 4),  ' % missing values')
```

## Numerical Features detection along with filling null values!

```python
numerical_features = [feature for feature in df.columns if df[feature].dtypes != 'O']

print('Number of numerical variables: ', len(numerical_features))
df[numerical_features].head()
```

```python
for feature in numerical_features:
    mean_value = df[feature].mean() 
    df[feature].fillna(mean_value, inplace=True)  

df[numerical_features].head()
```

```python
df[numerical_features].isna().sum()
```

```python
df.isna().sum()
```

## Finding Descrete columns and filling null values with it's mode: 

```python
df.dtypes
```

```python
object_features = [feature for feature in df.columns if df[feature].dtypes == 'O']
print('Number of object (categorical) variables: ', len(object_features))
df[object_features].head()
```

```python
for feature in object_features:
    mode_value = df[feature].mode()[0]  
    df[feature].fillna(mode_value, inplace=True)
df[object_features].head(5)
```
