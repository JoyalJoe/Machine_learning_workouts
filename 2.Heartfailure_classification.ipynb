# Standardization or normalization

```python
df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score
```

```python
x = df.drop(columns=['diabetes'], axis=1)
y  =  df['diabetes']
```

```python
x.shape
```

```python
y.shape
```

```python
scalar = StandardScaler()
standardized_data_x = scalar.fit_transform(x)
standardized_data_x.shape
```

```python
standardized_data_x
```

```python
x_train, x_test, y_train, y_test = train_test_split(standardized_data_x, y, test_size= 0.2, random_state= 3)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

# Model Development and Evaluation!

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
k_neighbors = KNeighborsClassifier(n_neighbors=5)
```

```python
k_neighbors.fit(x_train, y_train)
```

```python
train_predicted = k_neighbors.predict(x_train)
```

```python
training_acc = f1_score(train_predicted, y_train)
training_acc_score= accuracy_score(train_predicted, y_train)
```

```python
print(training_acc)
training_acc_score
```

```python
test_predicted = k_neighbors.predict(x_test)
testing_acc= f1_score(test_predicted, y_test)
testing_acc_score = accuracy_score(test_predicted, y_test)
```

```python
print(testing_acc)
testing_acc_score
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(x_train, y_train)
```

```python
train_predicted_lg = lg.predict(x_train)
```

```python
training_acc_lg = f1_score(train_predicted_lg, y_train)
training_acc_score_lg= accuracy_score(train_predicted_lg, y_train)
```

```python
print(training_acc_lg)
training_acc_score_lg
```

```python
test_predicted_lg = lg.predict(x_test)
testing_acc_lg= f1_score(test_predicted_lg, y_test)
testing_acc_score_lg = accuracy_score(test_predicted_lg, y_test)
```

```python
print(testing_acc_lg)
testing_acc_score_lg
```

```python
from sklearn.metrics import classification_report
```

```python
cls = classification_report(test_predicted_lg, y_test)
cls
```
